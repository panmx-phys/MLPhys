```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
dataPath = "data/"
beta_scan = np.arange(0.01,10,step = 0.1)
```

```python
res = np.load(dataPath+ "2022-03-24 22:47.npy")
jaxRes = np.load(dataPath+ "jax3.npy")
baseRes = np.load(dataPath+ "2022-03-25 12:42.npy")
```

```python
figure(figsize=(12, 6), dpi=80)
plt.scatter(beta_scan,res[:,-1] ,s = 10, label = "Diagonlization")
plt.scatter(beta_scan,jaxRes[:,0] ,s = 10, label = "Full Unitary (Jax)")
plt.scatter(beta_scan,baseRes[:,0] ,s = 10, label = "Quantum Computing without Noise")
plt.legend(loc = 'upper right')
plt.xlabel(r'$\beta $(1/T)')
plt.ylabel("Free Energy")
plt.title(r"Free Energy Optimization in Transverse Field Ising Model ($J =h = 0.5$)")
```

```python
figure(figsize=(12, 6), dpi=80)
plt.scatter(beta_scan,res[:,-1] ,s = 10, label = "Diagonlization")
plt.scatter(beta_scan,baseRes[:,0] ,s = 10, label = "Without Noise")
plt.scatter(beta_scan,res[:,0] ,s = 10, label = "With Noise")
plt.legend(loc = 'upper right')
plt.xlabel(r'$\beta $(1/T)')
plt.ylabel("Free Energy")
plt.title(r"Free Energy Optimization in Transverse Field Ising Model ($J =h = 0.5$)")
```
